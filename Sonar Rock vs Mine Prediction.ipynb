#### <center>1. Import the lib's</center>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### 7. Speratring the data and the label

```python
x= sonar_data.drop(columns=60, axis= 1)
y= sonar_data[60]
```

#### <center>3. Traning and testing data</center>

#### 1. Split the data

```python
x_train , x_test, y_train , y_test= train_test_split(x,y,test_size=0.1 ,
                                                     stratify=y, random_state=1)
```

#### <center>4. Trainning the model - Logistic Regression</center>

#### 1. Define model

```python
model = LogisticRegression()
```

#### 2. Training the logistic regression model with training data

```python
model.fit(x_train, y_train)
```

#### <center>5. Model Evalutation</center>

#### 1. Accuracy on our training data

```python
x_train_predicition = model.predict(x_train)
trainning_data_accuracy= accuracy_score(x_train_predicition, y_train)
```

#### 2. Print the accuracy on the training data

```python
print("Accuracy on training data", trainning_data_accuracy)
```

#### 3. Accuracy on test data

```python
x_test_predicition = model.predict(x_test)
test_data_accuracy= accuracy_score(x_test_predicition, y_test)
```

```python
print("Accuracy on training data", test_data_accuracy)
```
